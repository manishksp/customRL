```python
import pandas as pd
import random
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the CSV file
file_path = 'processed_data_notimefields.csv'
df = pd.read_csv(file_path)

# Selecting the relevant columns
columns_of_interest = [
    'conferenceDuration', 'conferenceCount', 'consultDuration', 'consultCount',
    'holdCount', 'holdDuration', 'connectedDuration', 'connectedCount',
    'wrapupDuration', 'ringingDuration'
]
df_selected = df[columns_of_interest]

# Derived Columns
df_selected['EngagementIntensity'] = df_selected['connectedDuration'] * df_selected['connectedCount']
df_selected['ConsultEfficiency'] = df_selected['consultDuration'] / df_selected['consultCount']
df_selected['HoldImpact'] = df_selected['holdDuration'] * df_selected['holdCount']
df_selected['WrapUpEfficiency'] = df_selected['wrapupDuration'] / (df_selected['conferenceCount'] + df_selected['consultCount'])

# Handling possible division by zero
df_selected['ConsultEfficiency'].replace(np.inf, 0, inplace=True)
df_selected['WrapUpEfficiency'].replace(np.inf, 0, inplace=True)

# Scaling the derived columns
scaler = MinMaxScaler()
df_selected[['EngagementIntensity', 'ConsultEfficiency', 'HoldImpact', 'WrapUpEfficiency']] = scaler.fit_transform(
    df_selected[['EngagementIntensity', 'ConsultEfficiency', 'HoldImpact', 'WrapUpEfficiency']]
)

# Display the updated dataframe
df_selected.head()

```

```python
# Environment to generate synthetic data and provide rewards
class Environment:
    def __init__(self):
        pass

    def step(self, action):
        """
        Simulate the environment's response to the agent's action.
        :param action: The action taken by the agent.
        :return: The next state and the reward.
        """
        connectedDuration = 0
        connectedCount = 0
        consultDuration = 0
        consultCount = 0
        holdDuration = 0
        holdCount = 0
        wrapupDuration = 0
        conferenceCount = 0
        ringingDuration = 0

        # Generate random data based on action
        if action == 0:  # Increase Engagement
            connectedDuration = random.uniform(1e5, 1e6)
            connectedCount = random.randint(1, 10)
            consultDuration = random.uniform(1e4, 1e5)
            consultCount = random.randint(1, 3)
            # Add holdCount for action 0
            holdDuration = random.uniform(1e3, 5e4)
            holdCount = random.randint(1, 3)
            conferenceCount =0
        elif action == 1:  # Improve Consult Efficiency
            consultDuration = random.uniform(1e4, 1e5)
            consultCount = random.randint(1, 5)
            connectedDuration = random.uniform(5e4, 2e5)
            connectedCount = random.randint(1, 5)
            # Add holdCount for action 1
            holdDuration = random.uniform(1e3, 5e4)
            holdCount = random.randint(1, 3)
            conferenceCount =0
        elif action == 2:  # Reduce Hold Impact
            holdDuration = random.uniform(1e3, 5e4)
            holdCount = random.randint(1, 3)
            connectedDuration = random.uniform(1e5, 5e5)
            connectedCount = random.randint(1, 5)
            conferenceCount =0
        elif action == 3:  # Optimize Wrap-Up
            wrapupDuration = random.uniform(1e3, 1e4)
            conferenceCount = random.randint(0, 2)
            consultCount = random.randint(1, 3)
            ringingDuration = random.uniform(500, 3000)
        else:
            # Default random generation for unhandled actions
            connectedDuration = random.uniform(1e5, 1e6)
            connectedCount = random.randint(1, 10)
            consultDuration = random.uniform(1e4, 1e5)
            consultCount = random.randint(1, 3)
            holdDuration = random.uniform(1e3, 5e4)
            holdCount = random.randint(1, 3)
            wrapupDuration = random.uniform(1e3, 1e4)
            conferenceCount = random.randint(0, 2)
            ringingDuration = random.uniform(500, 3000)

        # Generate the rest of the columns randomly
        conferenceDuration = random.uniform(1e3, 5e4)
        holdDuration = random.uniform(1e3, 5e4)
        wrapupDuration = random.uniform(1e3, 1e5)
        ringingDuration = random.uniform(500, 1e4)

        # Calculate derived columns
        engagement_intensity = connectedDuration * connectedCount
        consult_efficiency = consultDuration / max(consultCount, 1)
        hold_impact = holdDuration * holdCount
        wrap_up_efficiency = wrapupDuration / max(conferenceCount + consultCount, 1)

        # Define reward function (example: higher efficiency and lower impact gets higher rewards)
        reward = (
            1.0 / (1 + np.exp(-engagement_intensity))  # Reward for higher engagement
            + 1.0 / (1 + np.exp(-consult_efficiency))  # Reward for better consult efficiency
            + 1.0 / (1 + np.exp(-hold_impact))  # Reward for lower hold impact
            + 1.0 / (1 + np.exp(-wrap_up_efficiency))  # Reward for better wrap-up efficiency
        )

        # Return the next state and the reward
        next_state = (engagement_intensity, consult_efficiency, hold_impact, wrap_up_efficiency)
        generated_data = {
            'conferenceDuration': conferenceDuration,
            'conferenceCount': conferenceCount,
            'consultDuration': consultDuration,
            'consultCount': consultCount,
            'holdCount': holdCount,
            'holdDuration': holdDuration,
            'connectedDuration': connectedDuration,
            'connectedCount': connectedCount,
            'wrapupDuration': wrapupDuration,
            'ringingDuration': ringingDuration
        }
        return next_state, reward, generated_data

```

```python
# Q-Learning Agent
class QLearningAgent:
    def __init__(self, actions, alpha=0.1, gamma=0.9, epsilon=0.1):
        self.actions = actions
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.q_table = {}

    def choose_action(self, state):
        if random.uniform(0, 1) < self.epsilon:
            action = random.choice(self.actions)
        else:
            self.q_table.setdefault(state, np.zeros(len(self.actions)))
            action = np.argmax(self.q_table[state])
        return action

    def learn(self, state, action, reward, next_state):
        self.q_table.setdefault(state, np.zeros(len(self.actions)))
        self.q_table.setdefault(next_state, np.zeros(len(self.actions)))
        predict = self.q_table[state][action]
        target = reward + self.gamma * np.max(self.q_table[next_state])
        self.q_table[state][action] += self.alpha * (target - predict)
```

```python

# Initialize the environment and agent
actions = [0, 1, 2, 3]  # Action mappings
env = Environment()
agent = QLearningAgent(actions)

# Train the agent for 200 epochs and generate synthetic data
synthetic_data = []
for epoch in range(200):
    state = (0, 0, 0, 0)  # Initial state (arbitrary starting point)
    for _ in range(100):  # Each epoch consists of 100 steps
        action = agent.choose_action(state)
        next_state, reward, generated_data = env.step(action)
        agent.learn(state, action, reward, next_state)
        state = next_state
        synthetic_data.append(generated_data)

# Convert synthetic data to a DataFrame
synthetic_df = pd.DataFrame(synthetic_data)

# Sample 500 events from the synthetic data
sampled_synthetic_df = synthetic_df.sample(n=500, random_state=42)

# Display the sampled synthetic data
sampled_synthetic_df.head()
```
